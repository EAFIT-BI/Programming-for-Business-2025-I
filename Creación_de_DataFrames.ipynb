#Creación de *DataFrames*

Una DataFrame es el formato estructurado para gestión de los datos en `Python`.

La librería que gestiona este tipo de estructuras se llama `Pandas`.

```python
# Primero cargamos las librerías que necesitamos
# para el desarrollo del código

import pandas as pd #gestiona los DataFrames
```

```python
# Segundo, creamos una lista con la información
# de los estudiantes

estudiantes = [['Mónica', 'Bejarano', 'Contaduría', 16],
               ['Lina', 'Sepúlveda', 'Electrónica', 42],
               ['Maria Camila', 'Rivillas', 'Contaduría', 17],
               ['Laura Sofía', 'Guzman', 'Contaduría', 16],
               ['Sofía', 'Pérez', 'Contaduría', 16],
               ['Juliana', 'Carvajal', 'Contaduría', 17],
               ['Miguel Ángel', 'Monroy', 'Contaduría', 17],
               ['Sara Isabel', 'Bermúdez', 'Contaduría', 19],
               ['Sara Valentina', 'Pulido', 'Contaduría', 18],
               ['Paulina', 'Bedoya', 'Contaduría', 17],
               ['Sury', 'Salazar', 'Contaduría', 17],
               ['Samuel', 'Tamayo', 'Contaduría', 17],
               ['Paulina', 'Chavarriaga', 'Contaduría', 17],
               ['Salome', 'Mora', 'Contaduría', 19],
               ['Santiago', 'Perdomo', 'Administración', 22],
               ['Juan José', 'Giraldo', 'Administración', 24]]

```

```python
# Una vez construida la lista, podemos proceder a crear el DataFrame

df_estudiantes = pd.DataFrame(estudiantes)

# Visualizamos para comprobar
df_estudiantes
```

```python
# Para ponerle nombre a las columnas procedemos así:
df_estudiantes.columns = ['Nombre', 'Apellido', 'Carrera', 'Edad']

# Visualizamos para comprobar. En este caso solo visualizaremos
# los primeros 5 estudiantes
df_estudiantes.head(5)
```

```python

# Para seleccionar una sola columna, podemos hacerlo de dos maneras

# Forma 1
df_estudiantes['Carrera']
```

```python
# Forma 2
df_estudiantes.Carrera
```

```python
# Supongamos que queremos seleccionar dos columnas: Nombre, Apellido
df_estudiantes[['Nombre', 'Apellido']]
```

```python
# ¿Cómo podríamos crear una nueva columna llamada "Nombre completo" con la
# unión del Nombre y el Apellido?

df_estudiantes['Nombre completo'] = df_estudiantes['Nombre'] + ' ' + df_estudiantes['Apellido']
df_estudiantes
```

```python
# Ahora, vamos a remover una fila de la tabla

df_estudiantes = df_estudiantes.drop(1)

df_estudiantes.tail(5)
```
